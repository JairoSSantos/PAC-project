```python
edges_fft = {}
fft_func = lambda y: np.fft.fftshift(np.abs(np.fft.fft(y)))
for func, arr in edges.items():
    freqs = []
    ffts = []
    Fs = []
    for dim in [arr, arr.T]:
        freq = np.fft.fftshift(np.fft.fftfreq(dim.shape[1], 1))
        fft = np.apply_along_axis(fft_func, 1, dim)
        loc = freq > 0.015
        freq = freq[loc]
        fft = fft[:, loc]
        freqs.append(freq)
        ffts.append(fft)
        Fs.append([freq[y == y.max()][0] for y in fft])
    edges_fft[func] = {
        'freqs':freqs,
        'ffts':ffts,
        'Fs':Fs
    }
```

```python
fft_func = lambda y: np.fft.fftshift(np.abs(np.fft.fft(y)))
def pixel_scale_edge(img):
    arr = filters.sobel(img)
    fs = []
    for dim in [arr, arr.T]:
        freq = np.fft.fftshift(np.fft.fftfreq(dim.shape[1], 1))
        fft = np.apply_along_axis(fft_func, 1, dim)
        loc = freq > 0.015
        freq = freq[loc]
        fft = fft[:, loc]
        F = [freq[y == y.max()][0] for y in fft]
        fs.append(mode(F).mode[0])
        #f, counts = np.unique(np.round(F, 5), return_counts=True)
        #f[counts == np.sort(counts)[-1]][0]
    return fs

pixel_scale_edge(img)
```

```python
def grid(shape, fx, fy, wx, wy):
    field = np.ones(shape)
    tx, ty = int(1/fx), int(1/fy)
    for i in range(int(shape[0]*fx)):
        field[i*tx - wx//2 : i*tx + wx//2, :] = 0
    for j in range(int(shape[1]*fy)):
        field[:, j*ty - wy//2 : j*ty + wy//2] = 0
    return field
```

```python
import sympy as sp
from sympy.utilities.lambdify import lambdify
gamma, p = sp.symbols('gamma, p', positive=True, real=True)
n, m = sp.symbols('n, m', positive=True, integer=True)
x, y = sp.symbols('x, y', real=True)
```

```python
a0 = (sp.integrate(1, (x, -p, gamma*p))/p + sp.integrate(-1, (x, gamma*p, p))/p).simplify()
a0
```

```python
an = (sp.integrate(sp.cos(n*sp.pi*x/p), (x, -p, gamma*p))/p + sp.integrate(-sp.cos(n*sp.pi*x/p), (x, gamma*p, p))/p).simplify()
an
```

```python
bn = (sp.integrate(sp.sin(n*sp.pi*x/p), (x, -p, gamma*p))/p + sp.integrate(-sp.sin(n*sp.pi*x/p), (x, gamma*p, p))/p).simplify()
bn
```

```python
f = a0/2 + sp.Sum(an*sp.cos(n*sp.pi*x/p), (n, 1, m)) + sp.Sum(bn*sp.sin(n*sp.pi*x/p), (n, 1, m))
f
```

```python
f_num = lambdify((x, p, gamma, m), f, 'numpy')
```

```python
X = np.linspace(-5, 5, 500)
plt.ylim(-2, 2)
plt.plot(X, f_num(X, 1, 0.5, 10))
```

$$
F(x, \gamma, \varphi, p) = \left\{
    \begin{matrix}
        1, (x - \varphi) \le \gamma_xp\\
        -1, (x - \varphi) > \gamma_xp
    \end{matrix}
\right. \simeq \gamma + \sum_{n=1}^{m} \frac{2 \left(\left(-1\right)^{n} - \cos{\left(\pi \gamma n \right)}\right) \sin{\left(\frac{\pi n (x - \varphi)}{p} \right)}}{\pi n} + \sum_{n=1}^{m} \frac{2 \sin{\left(\pi \gamma n \right)} \cos{\left(\frac{\pi n (x - \varphi)}{p} \right)}}{\pi n}
$$

$$
Z_{ij} = F_{ij}\left(X_{ij} - \varphi_x, \gamma_x, p_y\right) + F_{ij}\left(Y_{ij} - \varphi_y, \gamma_y, p_y\right)
$$

$$
J = \sum_i\sum_j \left(I_{ij} - Z_{ij}\left(X_{ij}, Y_{ij}, \beta_k\right)\right)^2 \Longrightarrow \beta_k = (\gamma_x, \gamma_y, p_x, p_y, \varphi_x, \varphi_y)
$$

$$
\beta_k^{(t+1)} = \beta_k^{(t)} - \eta_k\left.\frac{\partial J}{\partial \beta_k}\right|_{\beta_k = \beta_k^{(t)}}
$$

$$
\frac{\partial J}{\partial \beta_k} = 2\sum_i\sum_j \left(Z_{ij}\left(X_{ij}, Y_{ij}, \beta_k\right) - I_{ij}\right)\frac{\partial Z_{ij}}{\partial \beta_k}
$$

```python
gamma_x, gamma_y, p_x, p_y, phi_x, phi_y = sp.symbols('gamma_x, gamma_y, p_x, p_y, phi_x, phi_y', real=True)
Z = f.subs({x:x - phi_x, gamma:gamma_x, p:p_x, m:2}) #, f.subs({x:y - phi_y, gamma:gamma_y, p:p_y, m:2})
```

```python
variables = [gamma_x, gamma_y, p_x, p_y, phi_x, phi_y]
dZdB = {var:Z.diff(var) for var in variables}
```

```python
Z_func = lambdify([x] + variables, Z)
Z_diffs = {k:lambdify([x] + variables, v) for k, v in dZdB.items()}
```

```python
def rescale(arr, i, f):
    return i + (arr - arr.min())/(arr.max() - arr.min()) * (f - i)
```

```python
X, Y = np.meshgrid(np.linspace(0, 1, 256), np.linspace(0, 1, 256))
Z_ij = rescale(Z_func(X, *np.random.random(len(variables))), -1, 1)
plt.imshow(Z_ij, cmap='gray')
```
