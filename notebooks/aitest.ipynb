```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from IPython.display import display
```

```python
np.random.randint(0, 10, 4)
```

```python
fig = plt.figure(figsize=(20, 15))
grid = GridSpec(4, 5, figure=fig)

fig.add_subplot(grid[:2, 2:])
ax = fig.add_subplot(grid[:2, :2])
ax.set_aspect('equal')
#ax.axis('off')
fig.add_subplot(grid[2:, :2]).set_aspect('equal')
for i in range(2, 4):
    for j in range(2, 5):
        fig.add_subplot(grid[i, j])
fig.tight_layout()
```
