```python
import aimodel
import data
```

```python
unet = aimodel.UNet('unet_teste', data.load_dataset(as_tensor=True)).build((1, 2))
```

```python
unet.summary()
```

```python
unet.compile(optimizer='adam', loss='binary_crossentropy', metrics=[aimodel.mape])
```

```python
unet.fit(10, 48, 3)
```
